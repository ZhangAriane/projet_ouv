```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
df = pd.read_csv('arbre_times.csv')

# 清洁列名
df.columns = df.columns.str.strip()

# 创建一个图表的列表，以便于分析不同的度量
metrics = [
    'tree_size', 
    'construction_time', 
    'compression_by_list_time', 
    'compression_by_tree_time',
    'space_complexity_by_list',
    'space_complexity_by_tree',
    'space_complexity_construction'
]

# 绘制每个度量的图表
for metric in metrics:
    plt.figure(figsize=(10, 6))
    plt.plot(df['bits'], df[metric])
    plt.title(f'{metric.replace("_", " ").title()} vs Bits')
    plt.xlabel('Bits')
    plt.ylabel(metric.replace("_", " ").title())
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('arbre_times.csv')

# Clean the column names
df.columns = df.columns.str.strip()

# Visualize construction time as it varies with the number of bits
plt.figure(figsize=(10, 5))
plt.plot(df['bits'], df['construction_time'],  label='Construction Time')

# Visualize the compression time for list and tree as it varies with the number of bits
plt.plot(df['bits'], df['compression_by_list_time'],  label='Compression Time by List')
plt.plot(df['bits'], df['compression_by_tree_time'],  label='Compression Time by Tree')

# Add legend
plt.legend()

# Add title and axis labels
plt.title('Time Complexity')
plt.xlabel('Number of Bits')
plt.ylabel('Time (seconds)')

# Show the plot
plt.tight_layout()
plt.show()

# Visualize space complexity as it varies with the number of bits
plt.figure(figsize=(10, 5))
plt.plot(df['bits'], df['space_complexity_by_list'], linestyle='-', color='blue', label='Space Complexity by List')
plt.plot(df['bits'], df['space_complexity_by_tree'], linestyle='--', color='red', label='Space Complexity by Tree')
plt.plot(df['bits'], df['space_complexity_construction'], linestyle='-.', color='orange', label='Space Complexity Construction')

# Add legend
plt.legend()

# Add title and axis labels
plt.title('Space Complexity')
plt.xlabel('Number of Bits')
plt.ylabel('Space Units')

# Show the plot
plt.tight_layout()
plt.show()
```
