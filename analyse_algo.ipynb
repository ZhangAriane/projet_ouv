```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
df = pd.read_csv('arbre_times.csv')
# Clean the column names
df.columns = df.columns.str.strip()
# 
metrics = ['Tree Size', 'Construction Time', 'Compression Time List', 'Compression Time Tree']

# 
for metric in metrics:
    plt.figure(figsize=(10, 6))
    plt.plot(df['Bits'], df[metric], marker='o')
    plt.title(f'{metric} vs Bits')
    plt.xlabel('Bits')
    plt.ylabel(metric)
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Construction Time'], df['Compression Time Tree'], label='Compression by Tree')
plt.scatter(df['Construction Time'], df['Compression Time List'], label='Compression by List')
plt.title('Construction Time vs Compression Time')
plt.xlabel('Construction Time')
plt.ylabel('Compression Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
df = pd.read_csv('arbre_times.csv')

# 
df.columns = [col.strip() for col in df.columns]

# 
plt.figure(figsize=(10, 5))
plt.plot(df['bits'], df['construction_time'], marker='o', label='Construction Time')

# 
plt.plot(df['bits'], df['compression_time_liste'], marker='s', label='Compression Time by List')
plt.plot(df['bits'], df['compression_time_arbre'], marker='^', label='Compression Time by Tree')

# 
plt.legend()

# 
plt.title('Time Complexity')
plt.xlabel('Number of Bits')
plt.ylabel('Time (seconds)')

# 
plt.tight_layout()
plt.show()

# 
plt.figure(figsize=(10, 5))
plt.plot(df['bits'], df['space_complexity_list'], marker='o', linestyle='-', color='blue', label='Space Complexity by List')
plt.plot(df['bits'], df['space_complexity_tree'], marker='s', linestyle='--', color='red', label='Space Complexity by Tree')
plt.plot(df['bits'], df['space_complexity_construction'], marker='^', linestyle='-.', color='orange', label='Space Complexity Construction')

# 
plt.legend()

# 
plt.title('Space Complexity')
plt.xlabel('Number of Bits')
plt.ylabel('Space Units')

# 
plt.tight_layout()
plt.show()
```
