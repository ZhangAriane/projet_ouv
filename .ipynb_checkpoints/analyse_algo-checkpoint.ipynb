```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data from CSV
df = pd.read_csv('arbre_times.csv')

# Clean the column names
df.columns = df.columns.str.strip()

# Use the DataFrame index as the x-axis for the plot
x_axis = df.index

# 绘图
plt.figure(figsize=(10, 5))

# Plot construction time
plt.plot(x_axis, df['Construction Time'], label='Construction Time', marker='o')

# Plot compression time list
plt.plot(x_axis, df['Compression Time List'], label='Compression Time List', marker='x')

# Plot compression time tree
plt.plot(x_axis, df['Compression Time Tree'], label='Compression Time Tree', marker='^')

# Add labels and legend
plt.xlabel('Data Point Index')
plt.ylabel('Time (seconds)')
plt.legend()
plt.title('Time Performance for Different Operations')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
df = pd.read_csv('arbre_times.csv')
# Clean the column names
df.columns = df.columns.str.strip()
# 创建一个图表的列表，以便于分析不同的度量
metrics = ['Tree Size', 'Construction Time', 'Compression Time List', 'Compression Time Tree']

# 绘制每个度量的图表
for metric in metrics:
    plt.figure(figsize=(10, 6))
    plt.plot(df['Bits'], df[metric], marker='o')
    plt.title(f'{metric} vs Bits')
    plt.xlabel('Bits')
    plt.ylabel(metric)
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Construction Time'], df['Compression Time Tree'], label='Compression by Tree')
plt.scatter(df['Construction Time'], df['Compression Time List'], label='Compression by List')
plt.title('Construction Time vs Compression Time')
plt.xlabel('Construction Time')
plt.ylabel('Compression Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
df = pd.read_csv('arbre_times.csv')
# Clean the column names
df.columns = df.columns.str.strip()
# 设置图表大小
plt.figure(figsize=(12, 8))

# 绘制构建时间曲线
plt.plot(df['Bits'], df['Construction Time'], marker='o', label='Construction Time')

# 绘制列表压缩时间曲线
plt.plot(df['Bits'], df['Compression Time List'], marker='s', label='Compression Time List')

# 绘制树压缩时间曲线
plt.plot(df['Bits'], df['Compression Time Tree'], marker='^', label='Compression Time Tree')

# 添加图例
plt.legend()

# 添加标题和轴标签
plt.title('Comparison of Construction and Compression Times')
plt.xlabel('Bits')
plt.ylabel('Time (seconds)')

# 显示网格
plt.grid(True)

# 显示图表
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
df = pd.read_csv('arbre_times.csv')

# 清洁列名
df.columns = [col.strip() for col in df.columns]

# 可视化树的构建时间随比特数的变化
plt.figure(figsize=(10, 5))
plt.plot(df['bits'], df['construction_time'], marker='o', label='Construction Time')

# 可视化树和列表压缩时间随比特数的变化
plt.plot(df['bits'], df['compression_time_liste'], marker='s', label='Compression Time by List')
plt.plot(df['bits'], df['compression_time_arbre'], marker='^', label='Compression Time by Tree')

# 添加图例
plt.legend()

# 添加标题和轴标签
plt.title('Time Complexity')
plt.xlabel('Number of Bits')
plt.ylabel('Time (seconds)')

# 显示图表
plt.tight_layout()
plt.show()

# 可视化空间复杂度随比特数的变化
plt.figure(figsize=(10, 5))
plt.plot(df['bits'], df['space_complexity_list'], marker='o', linestyle='-', color='blue', label='Space Complexity by List')
plt.plot(df['bits'], df['space_complexity_tree'], marker='s', linestyle='--', color='red', label='Space Complexity by Tree')
plt.plot(df['bits'], df['space_complexity_construction'], marker='^', linestyle='-.', color='orange', label='Space Complexity Construction')

# 添加图例
plt.legend()

# 添加标题和轴标签
plt.title('Space Complexity')
plt.xlabel('Number of Bits')
plt.ylabel('Space Units')

# 显示图表
plt.tight_layout()
plt.show()
```
